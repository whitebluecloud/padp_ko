# 3.9.2. General Infix-to-Postfix Conversion¶
# 3.9.2. 중위 표현법 후위 표현법으로 변환하기

* We need to develop an algorithm to convert any infix expression to a postfix expression. To do this we will look closer at the conversion process.
* 중위 표현법을 후위 표현법으로 변환하기 위한 알고리즘을 개발해야한다.
* 이것을 하기위해 변환 프로세스를 좀더 살펴보자.

* Consider once again the expression A + B * C. As shown above, A B C * + is the postfix equivalent. 
* A + B * C을 다시 한번 생각해보자 A B C * +의 후위 표현법과 똑같다.
* We have already noted that the operands A, B, and C stay in their relative positions. 
* 우리는 이미 A, B와 C 피연산자들과 그것들의 연관된 위치에 대해 생각해 보았다.
* It is only the operators that change position. 
* 연산자들만 위치를 변경하면 되는것이다
* Let’s look again at the operators in the infix expression. 
* 중위 표현법에서 연산자들을 한번 더 살펴보자.
* The first operator that appears from left to right is +. 
* 왼쪽에서 오른쪽으로 볼때 첫번째 연산자는 + 이다
* However, in the postfix expression, + is at the end since the next operator, *, has precedence over addition. 
* 그러나 후위 연산자 표기에서는 +가 * 뒤에 오게된다 왜냐하면 *가 더 우선이 되는 연산순위를 가지고 있기 때문.
* The order of the operators in the original expression is reversed in the resulting postfix expression.
* 최초 표현식에서 연산자들은 후위 표기법에서 역순으로 나타나있다.

* As we process the expression, the operators have to be saved somewhere since their corresponding right operands are not seen yet.
* 표현식을 진행하면서 연산자들은 일치하는 피연산자들이 아직 존재하지 않을때 저장되어야 한다.
* Also, the order of these saved operators may need to be reversed due to their precedence. 
* 또한 이 저장된 연산자들은 그들의 연산순서에 따라서 역으로 나타내야 한다.
* This is the case with the addition and the multiplication in this example. 
* 이것은 위 예제에서 덧셈과 곱셈의 예제로 나타난다.
* Since the addition operator comes before the multiplication operator and has lower precedence, it needs to appear after the multiplication operator is used.
* 덧셈 연산자가 곱셈 연산자 이전에 있지만, 연산 순위가 낮으므로 곱셈 연산자가 사용된 이후에 나타나야 한다.
* Because of this reversal of order, it makes sense to consider using a stack to keep the operators until they are needed.
* 이처럼 역순으로 계산되어야 하기 때문에, 스택을 사용하여 오퍼레이터들을 저장해놔야한다.

* What about (A + B) * C? Recall that A B + C * is the postfix equivalent. 
* (A + B) * C는 어떤가 A B + C * 가 후위 표현식이다.
* Again, processing this infix expression from left to right, we see + first. 
* 중위 표현식으로 나타낸 식에서, 왼쪽부터 오른쪽으로 봤을때 +가 먼저 보인다.
* In this case, when we see *, + has already been placed in the result expression because it has precedence over * by virtue of the parentheses. 
* 위 케이스에서는 +가 괄호 안에들어 있기때문에 표현식에서 *보다 앞서 나오게 된다.
* We can now start to see how the conversion algorithm will work. 
* 이제 우리는 변환 알고리즘을 시작할 수 있다
* When we see a left parenthesis, we will save it to denote that another operator of high precedence will be coming. 
* 왼쪽 괄호를 보게 된다면, 이것을 저장해 두고 이것보다 높은 연산순위의 것이 오는지 판단한다.
* That operator will need to wait until the corresponding right parenthesis appears to denote its position (recall the fully parenthesized technique). 
* 이 연산자는 오른쪽 괄호가 나올때 까지 기다리고, 일치하는 오른쪽 괄호가 나오면 위치를 표시한다.
* When that right parenthesis does appear, the operator can be popped from the stack.
* 오른쪽 괄호가 나온다면, 연산자는 스택에서 팝된다. 

* As we scan the infix expression from left to right, we will use a stack to keep the operators. 
* 중위 표현식을 왼쪽에서 오른쪽으로 스캔하면서, 이 연산자들을 저장하기 위해 스택을 사용할 것이다.
* This will provide the reversal that we noted in the first example. 
* 이는 첫 예제에서 역순을 사용하는 기능을 제공할 것이다.
* The top of the stack will always be the most recently saved operator. 
* 스태그이 맨 위는 가장 최근에 저장한 연산자일 것이다.
* Whenever we read a new operator, we will need to consider how that operator compares in precedence with the operators, if any, already on the stack.
* 새로운 연산자를 읽게되면, 이 연산자가 저장된 연산자보다 높은 연산순서를 가지고 있는지를 판단해야 한다.

* Assume the infix expression is a string of tokens delimited by spaces. 
* 중위 표현식이 스페이스로 구분된 문자열로 가정하자.
* The operator tokens are *, /, +, and -, along with the left and right parentheses, ( and ). 
* 연산자 토큰들은 소괄호와 * / + -이다.
* The operand tokens are the single-character identifiers A, B, C, and so on. 
* 피연산자들은 A,B,C 같은 한글자 이다
* The following steps will produce a string of tokens in postfix order.
* 아래는 후위 표현식을 나타내는 문자열을 반환할 것이다.

1. Create an empty stack called opstack for keeping operators. Create an empty list for output.
* op 스택이라고 불리는 빈 스택을 만들어 연산자들을 보관할수 있게 하라.
* 문자 출력용 빈 리스트를 생성하라
2. Convert the input infix string to a list by using the string method split.
* 스트링의 split메소드를 사용하여 중위 표현법을 LIST로 만들어라
3. Scan the token list from left to right.
* 리스트를 왼쪽에서 오른쪽 순서로 읽어라
  * If the token is an operand, append it to the end of the output list.
  * 만약 토큰이 피연산자라면, 출력 리스트에 APPEND하라
  * If the token is a left parenthesis, push it on the opstack.
  * 만약 왼쪽 괄호가 나온다면 OP스택에 푸쉬하라
  * If the token is a right parenthesis, pop the opstack until the corresponding left parenthesis is removed. Append each operator to the end of the output list.
  * 만약 오른쪽 괄호가 나온다면, opstack을 해당하는 왼쪽 괄호가 나올떄까지 pop 하라.
  * 각 연산자들을 아웃풋 리스트에 푸쉬해라
  * If the token is an operator, *, /, +, or -, push it on the opstack. However, first remove any operators already on the opstack that have higher or equal precedence and append them to the output list.
  * 만약 연산자들이 나온다면 (*, /, +, -) op스택에 푸쉬하라, 만약 op 스택에 맨 위 연산자가 더 높은 우선순위를 가지고 있다면 output list에 어펜드하라
4. When the input expression has been completely processed, check the opstack. Any operators still on the stack can be removed and appended to the end of the output list.
* 입력된 표현식이 모두 프로세싱되면, op스택을 체크하라. 만약 스택에 남아있는 연산자가 있다면 아웃풋 리스트에 POP하면서 어펜드할것

* Figure 9 shows the conversion algorithm working on the expression A * B + C * D. Note that the first * operator is removed upon seeing the + operator. 
* 9번 그림은 A * B + C * D를 변환하는 알고리즘을 보여준다.
* 처음 * 연산자가 + 오퍼레이터를 만나면 POP되는것을 잘 봐라
* Also, + stays on the stack when the second * occurs, since multiplication has precedence over addition. 
* 또한 + 은 두번째 *를 만나더라도, +는 *보다 연산순위가 낮기때문에 계속 스택에 남아있는다
* At the end of the infix expression the stack is popped twice, removing both operators and placing + as the last operator in the postfix expression.
* 중위 표현식의 마지막에서는 스택이 두번 POP되고 +가 후위 표현식의 마지막 연산자가 된다.

* In order to code the algorithm in Python, we will use a dictionary called prec to hold the precedence values for the operators. 
* 파이썬의 코드를 만들기 위해, PREC이라고 불리는 딕셔너리를 만들어, 연산자 우선순위를 체크할 것이다.
* This dictionary will map each operator to an integer that can be compared against the precedence levels of other operators (we have arbitrarily used the integers 3, 2, and 1). 
* 이 딕셔너리형은 각 연산자가 다른 연산 우선순서를 가지고 있는지 나타낸다. (명시적으로 숫자로 3,2,1로 사용함)
* The left parenthesis will receive the lowest value possible. 
* 왼쪽 괄호는 가장 낮은 값을 가진다.
* This way any operator that is compared against it will have higher precedence and will be placed on top of it. 
* 이것은 왼쪽 괄호와 비교하는 어떤 연산자도 더 높은 우선순위를 가지고있게 함으로써, 스택의 탑쪽에 계속 쌓이게 하려고하기 때문이다.
* Line 15 defines the operands to be any upper-case character or digit. The complete conversion function is shown in ActiveCode 1.
* 15라인은 피연산자가 숫자형이거나 문자형임을 나타낸다. 아래에 최종 코드가 있다.

````python
from pythonds.basic.stack import Stack

def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1
    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfixList.append(token)
        elif token == '(':
            opStack.push(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        else:
            while (not opStack.isEmpty()) and \
               (prec[opStack.peek()] >= prec[token]):
                  postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    return " ".join(postfixList)

print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
````

A few more examples of execution in the Python shell are shown below.

>>> infixtopostfix("( A + B ) * C")
'A B + C *'

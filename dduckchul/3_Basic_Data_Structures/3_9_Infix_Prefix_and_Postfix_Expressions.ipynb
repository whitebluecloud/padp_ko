# 3.9. Infix Prefix and Postfix Expressions
# 3.9. 중위, 전위, 후위 표기법

* When you write an arithmetic expression such as B * C, the form of the expression provides you with information so that you can interpret it correctly.
* 만약 당신이 B * C와 같은 수식을 작성한다면, 이 표기법의 형태는 당신이 올바르게 해석할수 있을 것이다.
* In this case we know that the variable B is being multiplied by the variable C since the multiplication operator * appears between them in the expression. 
* 이 경우에는 변수 B가 변수 C만큼 곱해진다는것을 곱셈 연산자 *가 변수들 가운데 있으므로 알수있다.
* This type of notation is referred to as infix since the operator is in between the two operands that it is working on.
* 두개의 피연산자 사이에 연산자가 있는 이런 형태의 표기법을 중위 표기법이라고 한다

* Consider another infix example, A + B * C. The operators + and * still appear between the operands, but there is a problem. 
* 다른 형태의 중위 표기법을 생각해보라, A+B*C 는 +와 * 연산자가 피연산자들 가운데 있다. 그러나 다른 문제점들이 있음
* Which operands do they work on? Does the + work on A and B or does the * take B and C? The expression seems ambiguous.
* 어떤 연산자를 어떻게 사용해야 하나? + 연산자가 A와 B를 먼저 받아야하나 * 연산자가 B와 C를 받아야 하는가? 이런 표기법은 애매해 보인다.

* In fact, you have been reading and writing these types of expressions for a long time and they do not cause you any problem. 
* 실은 당신이 이런 표기법을 오랜시간동안 쓰고 읽어왔기 떄문에 문제를 일으키진 않는다.
* The reason for this is that you know something about the operators + and *. 
* 문제를 일으키지 않는 이유는 당신이 + 연산자와 * 연산자에 대해 알기 떄문이다.
* Each operator has a precedence level. Operators of higher precedence are used before operators of lower precedence. 
* 각 연산자들은 수행 단계가 있고, 연산자들중에 높은 수행단계를 가진 연산자가 있고, 이것들은 낮은 단계보다 더 먼저 계산되어야 한다.
* The only thing that can change that order is the presence of parentheses. 
* 유일하게 이 연산 순서를 바꿀수 있는 것은 괄호가 나오는지 여부이다.
* The precedence order for arithmetic operators places multiplication and division above addition and subtraction. 
* *, /, +, - 보다 괄호가 들어있는 연산자가 더 먼저 수행되어야 한다
* If two operators of equal precedence appear, then a left-to-right ordering or associativity is used.
* 만약 두개의 연산자가 동일한 연산순서를 가지고 있다면 오니쪽에서 오른쪽으로 수행되어야 한다.

* Let’s interpret the troublesome expression A + B * C using operator precedence. 
* 이 문제 있는 표현법인 A+B*C를 연산자 우선순서를 사용하여 표기해 보자.
* B and C are multiplied first, and A is then added to that result. 
* B와 C가 먼저 곱해지고, A가 나중에 결과에 더해진다
* (A + B) * C would force the addition of A and B to be done first before the multiplication. 
* (A + B) * C는 A와 B의 덧셈을 먼저 수행하고, C가 그 이후에 곱해진다
* In expression A + B + C, by precedence (via associativity), the leftmost + would be done first.
* A+B+C의 표현에서는 가장 왼쪽의 +가 먼저 수행된다

* Although all this may be obvious to you, remember that computers need to know exactly what operators to perform and in what order. 
* 비록 이것들이 당신에게는 당연하게 보일지 모르겠으나, 컴퓨터는 정확하게 무슨 연산자가 어떻게 연산되어야 하는지 알아야 한다.
* One way to write an expression that guarantees there will be no confusion with respect to the order of operations is to create what is called a fully parenthesized expression. 
* 이런 표현법을 헷갈림 없이 완전하게 동일 순서대로 보장하는 방법중에는 모든것을 괄호 표현법으로 표기하는 방법이 있다.
* This type of expression uses one pair of parentheses for each operator. The parentheses dictate the order of operations; there is no ambiguity. There is also no need to remember any precedence rules.
* 이런 형태의 표기법은 각 연산자마다 괄호쌍 표기를 필요로 한다. 이 괄호들은 연산자의 순서를 명시한다.
* 이 표현법에는 애매함이 없고, 연산자의 규칙을 알 필요도 없다.

* The expression A + B * C + D can be rewritten as ((A + (B * C)) + D) to show that the multiplication happens first, followed by the leftmost addition. 
* A + B * C + D 표현법은 ((A + (B * C)) + D)로 다시 쓰여질수 있고, 이것은 곱셈이 가장 먼저 수행되고, 왼쪽에 가장 가까운 덧셈이 다음으로 계산된다는것을 알 수 있다.
* A + B + C + D can be written as (((A + B) + C) + D) since the addition operations associate from left to right.
* A + B + C + D 는 (((A + B) + C) + D)로 표현될 수 있고 덧셈 연산자는 왼쪽에서부터 오른쪽으로 연산된다

* There are two other very important expression formats that may not seem obvious to you at first. 
* 이제 당신이 보기에 명확해 보이지 않는 두개의 매우 중요한 연산자 표현법을 보여줄 것이다
* Consider the infix expression A + B. What would happen if we moved the operator before the two operands? 
* 중위 표현법인 A+B를 생각해 보라, 만약 두개의 피연산자들보다 앞에 연산자를 쓰면 어떻게되나..?
* The resulting expression would be + A B. Likewise, we could move the operator to the end. We would get A B +. These look a bit strange.
* 연산자의 표기법은 + A B 가 될것이다 마찬가지로 연산자를 끝으로 옮기면 A B +가 될것이다.

* These changes to the position of the operator with respect to the operands create two new expression formats, prefix and postfix.
* 이렇게 연산자들의 위치를 피연산자에 바꾸는 것들이 두개의 새로운 표현법을 나타내 준다. 전위 표기법과 후위 표현법이다.
* Prefix expression notation requires that all operators precede the two operands that they work on. 
* 전위 표현법은 모든 연산자들이 동작하기 위해 이후에 두개의 피연산자가 필요함을 나타낸다.
* Postfix, on the other hand, requires that its operators come after the corresponding operands. A few more examples should help to make this a bit clearer (see Table 2).
* 반대로 후위 표현법은 피연산자에 대응하는 연산자들이 그 이후에 나타나야 한다.
* 몇개 예제가 이것을 더 상세히 보여질 것이다.

* A + B * C would be written as + A * B C in prefix. The multiplication operator comes immediately before the operands B and C, denoting that * has precedence over +. 
* A+B*C는 전위 표현법으로 + A * B C 로 나타낸다. 곱셈 연산자가 피 연산자 B,C 보다 먼저 나타나고, 이것은 +보다 * 연산자가 더 우선순위를 가지고 있는것을 보여준다.
* The addition operator then appears before the A and the result of the multiplication.
* 더하기 연산자는 A와 곱셈의 결과를 더하면 된다

* In postfix, the expression would be A B C * +. Again, the order of operations is preserved since the * appears immediately after the B and the C, denoting that * has precedence, with + coming after.
* 후위 표현법으로는 A B C * + 가 된다. 마찬가지로 연산자의 순서는 B와 C이후로 *가 바로 나타나는데, 이것은 이후에 오는 + 보다 연산 순위가 높다는것을 나타낸다.
* Although the operators moved and now appear either before or after their respective operands, the order of the operands stayed exactly the same relative to one another.
* 비록 연산자들이 피연산자들보다 이전혹은 이후에 올지라도, 연산자들의 연산순서는 다른것들과 다르지 않다.

| Infix Expression | Prefix Expression | Postfix Expression |
|------------------|-------------------|--------------------|
| A + B            | + A B             | A B +              |
| A + B * C        | + A * B C         | A B C * +          |

* Now consider the infix expression (A + B) * C.
* 이제 중위 표기법 (A + B) * C를 생각해 보아라.
* Recall that in this case, infix requires the parentheses to force the performance of the addition before the multiplication. 
* 이 경우에는 괄호가 덧셈의 연산순서를 더 높여준 다는것을 기억하고 있어라.
* However, when A + B was written in prefix, the addition operator was simply moved before the operands, + A B. The result of this operation becomes the first operand for the multiplication.
* A + B가 전위 표기법으로 사용될떄 덧셈 연산자는 피 연산자의 앞으로 이동해 + A B가 되고, 그 앞에 곱셈 연산자가 처음 연산자로 오게 된다.
* The multiplication operator is moved in front of the entire expression, giving us * + A B C. Likewise, in postfix A B + forces the addition to happen first.
* 곱셈 연산자가 전체 표현식의 맨 앞에 오게되어서 * + A B C로 나타난다. 마찬가지로 후위 연산자 에서는 A B + 가 먼저 오게되어서 덧셈이 먼저 일어난다는 것을 알 수 있다.
* The multiplication can be done to that result and the remaining operand C. The proper postfix expression is then A B + C *.
* 곱셈이 그 이후에 일어나서, 덧셈의 결과와 C가 곱해진다. 후위 표현법의 적절한 표현식은 A B + C *.가 된다.

* Consider these three expressions again (see Table 3). Something very important has happened. Where did the parentheses go? Why don’t we need them in prefix and postfix? 
* 이 세가지 표현법을 다시 한번 봐라 (TABLE 3) 매우 중요한점이 보일것이다. 괄호는 어디간거야? 전위 표현법이나 후위 표현법에서는 이걸 안써도 되나?
* The answer is that the operators are no longer ambiguous with respect to the operands that they work on. Only infix notation requires the additional symbols. 
* 이것의 답은 연산자들이 순서에 따라 더이상 어떻게 동작하는지 애매하게 되지 않는 것이다.
* 오직 중위 표기법에서만 이런 추가적인 기호가 사용된다
* The order of operations within prefix and postfix expressions is completely determined by the position of the operator and nothing else. In many ways, this makes infix the least desirable notation to use.
* prefix와 postfix의 연산 순서는 연산자의 위치에만 달렸을 뿐이다. 이것은 중위 표현법을 가장 덜 사용하게 되는 이유이다.

| Infix Expression | Prefix Expression | Postfix Expression |
|------------------|-------------------|--------------------|
| (A + B) * C      | * + A B C         | A B + C *          |

* Table 4 shows some additional examples of infix expressions and the equivalent prefix and postfix expressions.
* 4번 표는 이런 중위 표현법과 그에 대응하는 전위, 후위표현법을 보여준다
* Be sure that you understand how they are equivalent in terms of the order of the operations being performed.
* 이 표현법이 연산 순서가 똑같다는것을 이해해라

| Infix Expression  | Prefix Expression | Postfix Expression |
|-------------------|-------------------|--------------------|
| A + B * C + D     | + + A * B C D     | A B C * + D +      |
| (A + B) * (C + D) | * + A B + C D     | A B + C D + *      |
| A * B + C * D     | + * A B * C D     | A B * C D * +      |
| A + B + C + D     | + + + A B C D     | A B + C + D +      |

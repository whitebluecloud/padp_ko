* Here is the procedure:
* 여기 그 절차가 있다 : 

* From our starting position we will first try going North one square and then recursively try our procedure from there.
* 시작 지점에서 시작해서, 먼저 북쪽 방향으로 이동하고, 재귀적으로 이 절차를 반복할 것이다.
* If we are not successful by trying a Northern path as the first step then we will take a step to the South and recursively repeat our procedure.
* 만약 북쪽으로 이동하는것을 실패한다면, 남쪽으로 방향을 이동하고 우리의 절차를 반복할 것이다.
* If South does not work then we will try a step to the West as our first step and recursively apply our procedure.
* 남쪽도 실패한다면 서쪽으로 이동하고 절차를 반복하자.
* If North, South, and West have not been successful then apply the procedure recursively from a position one step to our East.
* 만약 북쪽, 남쪽, 서쪽도 실패한다면 동쪽으로 이동하고 절차를 반복하자.
* If none of these directions works then there is no way to get out of the maze and we fail.
* 만약 이런 방향들중에 어떤것도 성공하지 못한다면, 미로를 벗어날 방법이 없는 것이므로 실패한 것이다.

* Now, that sounds pretty easy, but there are a couple of details to talk about first.
* 이제 이 문제가 쉬워보이긴 하지만, 첫번째 단계에서 자세한 점을 살펴보자.
* Suppose we take our first recursive step by going North. 
* 첫번째 재귀 호출인 북쪽으로 이동하는 것을 생각해보자.
* By following our procedure our next step would also be to the North. 
* 다음 반복하는 절차도 북쪽 이동이 될것이다.
* But if the North is blocked by a wall we must look at the next step of the procedure and try going to the South. 
* 하지만 만약 북쪽이 벽으로 막혀있다면 다음 단계를 실행하고, 남쪽으로 이동할 것이다.
* Unfortunately that step to the south brings us right back to our original starting place. 
* 불행히도 남쪽으로 이동하는 것은 원래 시작한 곳으로 되돌아 가게 할것이다.
* If we apply the recursive procedure from there we will just go back one step to the North and be in an infinite loop. 
* 만약 ㅇ리가 재귀 절차를 이렇게 적용한다면 한번 북쪽으로 갔다가 돌아가는 무한 루프에 빠질 것이다.
* So, we must have a strategy to remember where we have been. 
* 그러므로,우리가 어디를 방문 했었는지에 대해 기억하는 전력이 필요하다
* In this case we will assume that we have a bag of bread crumbs we can drop along our way. 
* 이 경우에서는 우리가 가는 길마다 빵조각을 놓기로 하자.
* If we take a step in a certain direction and find that there is a bread crumb already on that square, we know that we should immediately back up and try the next direction in our procedure.
* 만약 어떤 방향으로 움직이고, 그 사각형에서 빵조각을 발견한다면, 즉시 돌아가서 프로시져에서 다음을 수행해야 한다.
* As we will see when we look at the code for this algorithm, backing up is as simple as returning from a recursive function call.
* 우리가 이 알고리즘의 코드를 봤을때 돌아가는것은 재귀 호출하는 것만큼 간단할 것이다.

* As we do for all recursive algorithms let us review the base cases. 
* 모든 재귀 알고리즘을 구현하는 것처럼 base case를 살펴보자 
* Some of them you may already have guessed based on the description in the previous paragraph. 
* 아마 이미 몇몇은 이전 문단의 조건으로 base case를 추측할수 있었을 것이다.
* In this algorithm, there are four base cases to consider:
* 이 알고리즘에서 4가지 base case를 생각할 수 있다.

1. The turtle has run into a wall. Since the square is occupied by a wall no further exploration can take place.
* turtle이 벽으로 향했을때, 사각형이 벽으로 되어있다면 더이상 탐험이 계속될수 없을 것이다.
2. The turtle has found a square that has already been explored. We do not want to continue exploring from this position or we will get into a loop.
* turtle이 이미 방문한 사각형을 찾았을때. 우리는 이 위치에서 탐험을 하지 않을것이다 안그러면 무한루프에 빠질것임
3. We have found an outside edge, not occupied by a wall. In other words we have found an exit from the maze.
* 벽에 둘러싸지지 않은 가장자리 부분을 찾았을때. 다른말로 한다면 우리는 미로의 출구를 찾은 것이다.
4. we have explored a square unsuccessfully in all four directions.
* 우리가 모든 네 방향에 탐험을 실패하는것

* Let’s examine the code for the search function which we call searchFrom. 
* 이제 searchFrom 으로 불리는 search function 코드를 살펴보자
* The code is shown in Listing 3. 
* 이 코드는 Listing3 에서 볼 수 있다
* Notice that this function takes three parameters: a maze object, the starting row, and the starting column.
* 이 펑션이 세가지 파라미터를 받는것을 볼 수 있다. maze object, 시작 row, 시작 column
* This is important because as a recursive function the search logically starts again with each recursive call.
* 이것은 재귀적 호출로 반복할떄마다 어디서 시작하는지 표현하기 위해 중요하다.

```python
# Listing 3

def searchFrom(maze, startRow, startColumn):
    maze.updatePosition(startRow, startColumn)
   #  Check for base cases:
   #  1. We have run into an obstacle, return false
    if maze[startRow][startColumn] == OBSTACLE :
        return False
    #  2. We have found a square that has already been explored
    if maze[startRow][startColumn] == TRIED:
        return False
    # 3. Success, an outside edge not occupied by an obstacle
    if maze.isExit(startRow,startColumn):
        maze.updatePosition(startRow, startColumn, PART_OF_PATH)
        return True
    maze.updatePosition(startRow, startColumn, TRIED)

    # Otherwise, use logical short circuiting to try each
    # direction in turn (if needed)
    found = searchFrom(maze, startRow-1, startColumn) or \
            searchFrom(maze, startRow+1, startColumn) or \
            searchFrom(maze, startRow, startColumn-1) or \
            searchFrom(maze, startRow, startColumn+1)
    if found:
        maze.updatePosition(startRow, startColumn, PART_OF_PATH)
    else:
        maze.updatePosition(startRow, startColumn, DEAD_END)
    return found
```

* As you look through the algorithm you will see that the first thing the code does (line 2) is call updatePosition.
* 알고리즘을 살펴보면, 2라인에서 먼저 보이는 것은 updatePosition을 호출하는 것이다.
* This is simply to help you visualize the algorithm so that you can watch exactly how the turtle explores its way through the maze. 
* 이것은 turtle이 어떻게 미로를 탐험하는지에 대해 정확히 이동하는 것을 볼수 있으므로, 알고리즘을 시각화하는데 도움을 준다.
* Next the algorithm checks for the first three of the four base cases: Has the turtle run into a wall (line 5)? 
* 다음 알고리즘 확인은 4가지 base case중 처음 3가지를 확인하는 것이다. turtle이 벽으로 이동했는가?
* Has the turtle circled back to a square already explored (line 8)? 
* turtle이 이미 탐험했던 영역으로 되돌아갔는가?
* Has the turtle found an exit (line 11)? 
* turtle이 출구를 찾았는가?
* If none of these conditions is true then we continue the search recursively.
* 이 조건들중 어느것도 만족시키지 않았다면 재귀 호출을 계속 한다.

* You will notice that in the recursive step there are four recursive calls to searchFrom. 
* 또한 searchFrom에서 4번의 재귀 호출 단계를 확인할 수 있다.
* It is hard to predict how many of these recursive calls will be used since they are all connected by or statements. 
* 이 재귀 호출들이 얼마나 호출될지 예상하기는 힘들다.
* If the first call to searchFrom returns True then none of the last three calls would be needed. 
* 만약 searchFrom의 맨 첫번째가 true를 반환한다면, 나머지 세가지 call은 불필요 할 것이다.
* You can interpret this as meaning that a step to (row-1,column) (or North if you want to think geographically) is on the path leading out of the maze.
* 이것을 해석하자면 북쪽에 (row-1, column)에 미로를 탈출하는 길이 있다고 생각할 수 있다.
* If there is not a good path leading out of the maze to the North then the next recursive call is tried, this one to the South. 
* 만약 북쪽으로 이어지는 좋은 경로가 없다면, 다음 재귀 호출이 일어나고 이것은 남쪽으로 갈 것이다.
* If South fails then try West, and finally East. 
* 남쪽이 실패하면 서쪽, 마지막으로는 동쪽으로 이동할 것이다.
* If all four recursive calls return false then we have found a dead end.
* 만약 내가지 재귀 호출이 false를 반환하면, 우리는 막다른 길에 다다른것이다.
* You should download or type in the whole program and experiment with it by changing the order of these calls.

* Finally, the isExit method uses the current position of the turtle to test for an exit condition. 
* 마지막으로 isExit 메소드를 사용해 현재 turtle의 위치가 exit조건에 있는지 확인한다
* An exit condition is whenever the turtle has navigated to the edge of the maze, either row zero or column zero, or the far right column or the bottom row.

```python
import turtle

PART_OF_PATH = 'O'
TRIED = '.'
OBSTACLE = '+'
DEAD_END = '-'

class Maze:
    def __init__(self,mazeFileName):
        rowsInMaze = 0
        columnsInMaze = 0
        self.mazelist = []
        mazeFile = open(mazeFileName,'r')
        rowsInMaze = 0
        for line in mazeFile:
            rowList = []
            col = 0
            for ch in line[:-1]:
                rowList.append(ch)
                if ch == 'S':
                    self.startRow = rowsInMaze
                    self.startCol = col
                col = col + 1
            rowsInMaze = rowsInMaze + 1
            self.mazelist.append(rowList)
            columnsInMaze = len(rowList)

        self.rowsInMaze = rowsInMaze
        self.columnsInMaze = columnsInMaze
        self.xTranslate = -columnsInMaze/2
        self.yTranslate = rowsInMaze/2
        self.t = turtle.Turtle()
        self.t.shape('turtle')
        self.wn = turtle.Screen()
        self.wn.setworldcoordinates(-(columnsInMaze-1)/2-.5,-(rowsInMaze-1)/2-.5,(columnsInMaze-1)/2+.5,(rowsInMaze-1)/2+.5)

    def drawMaze(self):
        self.t.speed(10)
        self.wn.tracer(0)
        for y in range(self.rowsInMaze):
            for x in range(self.columnsInMaze):
                if self.mazelist[y][x] == OBSTACLE:
                    self.drawCenteredBox(x+self.xTranslate,-y+self.yTranslate,'orange')
        self.t.color('black')
        self.t.fillcolor('blue')
        self.wn.update()
        self.wn.tracer(1)

    def drawCenteredBox(self,x,y,color):
        self.t.up()
        self.t.goto(x-.5,y-.5)
        self.t.color(color)
        self.t.fillcolor(color)
        self.t.setheading(90)
        self.t.down()
        self.t.begin_fill()
        for i in range(4):
            self.t.forward(1)
            self.t.right(90)
        self.t.end_fill()

    def moveTurtle(self,x,y):
        self.t.up()
        self.t.setheading(self.t.towards(x+self.xTranslate,-y+self.yTranslate))
        self.t.goto(x+self.xTranslate,-y+self.yTranslate)

    def dropBreadcrumb(self,color):
        self.t.dot(10,color)

    def updatePosition(self,row,col,val=None):
        if val:
            self.mazelist[row][col] = val
        self.moveTurtle(col,row)

        if val == PART_OF_PATH:
            color = 'green'
        elif val == OBSTACLE:
            color = 'red'
        elif val == TRIED:
            color = 'black'
        elif val == DEAD_END:
            color = 'red'
        else:
            color = None

        if color:
            self.dropBreadcrumb(color)

    def isExit(self,row,col):
        return (row == 0 or
                row == self.rowsInMaze-1 or
                col == 0 or
                col == self.columnsInMaze-1 )

    def __getitem__(self,idx):
        return self.mazelist[idx]


def searchFrom(maze, startRow, startColumn):
    # try each of four directions from this point until we find a way out.
    # base Case return values:
    #  1. We have run into an obstacle, return false
    maze.updatePosition(startRow, startColumn)
    if maze[startRow][startColumn] == OBSTACLE :
        return False
    #  2. We have found a square that has already been explored
    if maze[startRow][startColumn] == TRIED or maze[startRow][startColumn] == DEAD_END:
        return False
    # 3. We have found an outside edge not occupied by an obstacle
    if maze.isExit(startRow,startColumn):
        maze.updatePosition(startRow, startColumn, PART_OF_PATH)
        return True
    maze.updatePosition(startRow, startColumn, TRIED)
    # Otherwise, use logical short circuiting to try each direction
    # in turn (if needed)
    found = searchFrom(maze, startRow-1, startColumn) or \
            searchFrom(maze, startRow+1, startColumn) or \
            searchFrom(maze, startRow, startColumn-1) or \
            searchFrom(maze, startRow, startColumn+1)
    if found:
        maze.updatePosition(startRow, startColumn, PART_OF_PATH)
    else:
        maze.updatePosition(startRow, startColumn, DEAD_END)
    return found


myMaze = Maze('maze2.txt')
myMaze.drawMaze()
myMaze.updatePosition(myMaze.startRow,myMaze.startCol)

searchFrom(myMaze, myMaze.startRow, myMaze.startCol)
```

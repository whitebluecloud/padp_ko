## 6.3. Vocabulary and Definitions

* Now that we have looked at examples of trees, we will formally define a tree and its components.
* 앞서 트리의 예제를 봤으므로, 이제 일반적인 트리의 정의와 요소들을 볼것이다.

### Node
* A node is a fundamental part of a tree. 
* 노드는 트리의 기초적인 부분이다.
* It can have a name, which we call the “key.” A node may also have additional information. 
* 이름을 가질 수 있고 (key로 부른다) 노드는 추가적인 정보를 가질 수도있다.
* We call this additional information the “payload.” While the payload information is not central to many tree algorithms, it is often critical in applications that make use of trees.
* 이 추가적인 정보들을 페이로드라고 부른다. 페이로드는 많은 트리 알고리즘에서 중요것은 아니지만, 트리를 사용하는 어플리케이션에서는 종종 중요한 것이 될수 있다.

### Edge
* An edge is another fundamental part of a tree.
* 엣지는 트리의 다른 기초적인 부분이다.
* An edge connects two nodes to show that there is a relationship between them. Every node (except the root) is connected by exactly one incoming edge from another node.
* 엣지는 두개의 노드를 잇고 두 노드 간의 관계가 있다는것을 표현한다. 루트를 제외한 모든 노드는 정확히 하나의 다른 노드로부터 들어오는 엣지가 있다.  
* Each node may have several outgoing edges.
* 각각의 노드는 여러개의 나가는 엣지가 있을 수 있다.

### Root
* The root of the tree is the only node in the tree that has no incoming edges.
* 루트는 트리에서 유일하기 들어오는 엣지가 없는 노드이다.
* In Figure Figure 2, / is the root of the tree.
* Figure 2에서 / 가 트리의 루트이다.

### Path
* A path is an ordered list of nodes that are connected by edges. 
* 경로는 엣지로 연결되어있는 노드들의 정렬된 리스트 이다.
* For example, Mammal → Carnivora → Felidae → Felis → Domestica is a path.
* 예를들면, 포유강 식육목 고양잇과 고양잇속 집고양이 가 패스이다

### Children
* The set of nodes c that have incoming edges from the same node to are said to be the children of that node.
* 들어오는 엣지의 노드가 같은 노드들의 집합을 자식들이라고 한다.
* In Figure Figure 2, nodes log/, spool/, and yp/ are the children of node var/.
* Figure2 에서 log/, spool/, yp/는 var/의 자식들이다.

### Parent
* A node is the parent of all the nodes it connects to with outgoing edges.
* 나가는 엣지로 연결된것들의 부모를 부모라고 한다.
* In Figure 2 the node var/ is the parent of nodes log/, spool/, and yp/.
* Figure 2는 var/ 노드가 log/, spool/, yp/의 부모이다.

### Sibling
* Nodes in the tree that are children of the same parent are said to be siblings. 
* 같은 부모를 가진 자식들끼리는 Sibling이라고 부른다.
* The nodes etc/ and usr/ are siblings in the filesystem tree.
* etc/와 usr/노드는 파일 시스템 트리에서 형제관계이다.

### Subtree
* A subtree is a set of nodes and edges comprised of a parent and all the descendants of that parent.
* subtree는 부모와 그것들을 구성하는 노드와 엣지로 이뤄진 것들이다.

### Leaf Node
* A leaf node is a node that has no children. 
* 리프 노드는 자식이 없는 노드이다.
* For example, Human and Chimpanzee are leaf nodes in Figure 1.
* 예를들어, 인간과 침팬지가 리프 노드이다.

### Level
* The level of a node n is the number of edges on the path from the root node to n.
* n 노드의 level은 루트부터 n 노드로 이어지는 경로의 엣지 갯수이다.
* For example, the level of the Felis node in Figure 1 is five.
* 예를들어, felis의 레벨은 5이다
* By definition, the level of the root node is zero.
* 루트 노드의 레벨은 0이다.

### Height
* The height of a tree is equal to the maximum level of any node in the tree.
* 트리의 높이는 트리에서 최대 레벨을 가진 노드와 같다.
* The height of the tree in Figure 2 is two.
* Figure2에서 트리의 높이는 2이다.

* With the basic vocabulary now defined, we can move on to a formal definition of a tree. 
* 기본적인 어휘들을 알아봤으므로, 트리의 일반적인 정의를 살펴보자.
* In fact, we will provide two definitions of a tree. 
* 트리의 두가지 정의를 살펴보자.
* One definition involves nodes and edges.
* 한 정의는 노드와 엣지로 구성되어있다.
* The second definition, which will prove to be very useful, is a recursive definition.
* 두번째 정의는 재귀적 정의이다. 이것은 매우 유용할것이다.

* Definition One: A tree consists of a set of nodes and a set of edges that connect pairs of nodes.
* 정의 1 : 트리는 노드들과 노드들의 쌍을 연결하는 엣지로 구성되어있다.

* A tree has the following properties:
* 트리는 다음과 같은 속성을 갖고 있다.

* One node of the tree is designated as the root node.
* 트리의 한개 노드는 루트 노드로 정해져있다.
* Every node n, except the root node, is connected by an edge from exactly one other node p, where p is the parent of n.
* 루트 노드를 제외한 모든 노드 n 들은 다른 하나의 노드 p와 엣지로 연결되어있다. (p는 n의 부모)
* A unique path traverses from the root to each node.
* 유니크한 경로들이 루트에서 각 노드로 순회한다.
* If each node in the tree has a maximum of two children, we say that the tree is a binary tree.
* 만약 각각의 노드들이 최대 두개의 자식으로 구성되어 있다면, 우리는 이것을 이진 트리라고 한다.

* Figure 3 illustrates a tree that fits definition one.
* Figure 3은 정의 1과 부합하는 트리를 보여주고 있다.
* The arrowheads on the edges indicate the direction of the connection.
* 엣지의 화살표들이 연결의 방향을 보여주고 있다.

* Definition Two: A tree is either empty or consists of a root and zero or more subtrees, each of which is also a tree.
* 정의 2 : 비어있거나, 루트와 0개 혹은 그 이상의 서브트리로 구성되어 있는 것들 또한 트리이다.
* The root of each subtree is connected to the root of the parent tree by an edge. 
* 각 서브트리의 루트는 부모 트리와 엣지로 연결되어 있다.
* Figure 4 illustrates this recursive definition of a tree. 
* Figure 4는 트리의 재귀적 정의를 나타내고 있다.
* Using the recursive definition of a tree, we know that the tree in Figure 4 has at least four nodes, since each of the triangles representing a subtree must have a root.
* 재귀적 정의의 트리를 사용해 Figure 4에 나타나 있는 트리가 최소 4개의 노드로 구성되어 있고, 각 삼각형들은 루트를 가지고 있는 subtree를 표현한다는 것을 알 수 있다.
* It may have many more nodes than that, but we do not know unless we look deeper into the tree.
* 이것은 이것보다 더 많은 노드를 가질 수 있으나, 트리의 안을 살펴보기 전까지는 알수 없다.

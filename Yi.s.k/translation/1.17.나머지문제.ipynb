```python
# 14. Design a class to represent a playing card. Now design a class to represent a deck of cards. Using these two classes, implement a favorite card game.
# 섰다 구현하겠습니다..... 암행어사 땡잡이 멍구사 구사 특수족보 제외하고 순수 카드 강함으로만...
import random
class flowerFight:
    def __init__(self, n):
        self.name = n
        self.output = None
        
    def getName(self):
        return self.name
    
    def getResult(self):
        self.getPick2()
        self.output = self.checkMyStat()
        return self.name+"의 패 : "+self.output
    
class deck(flowerFight):
    def __init__(self, n):
        flowerFight.__init__(self,n)
        self.pick1 = 0
        self.pick2 = 0
    
    def getPick1(self):
        if (self.pick1 == 0):
            self.pick1 += random.randrange(1,21)
            print(self.pick1)
        return self.pick1
    
    def getPick2(self):
        if (self.pick2 == 0):
            self.pick2 += random.randrange(1,21)
        return self.pick2

    def betting(self):
        self.betting = int(input(self.name+"님, 게임에 베팅하시겠습니까? [다이=1/콜=2/하프=3/쿼터=4/따당=5/올인=6]"))
        if(self.betting == 1):
            print(self.name+"님이 게임을 포기했습니다.")
            return str(self.betting)
        else:
            return str(self.betting)
    
    def checkMyStat(self):
        if(self.pick1%10 == self.pick2%10):
            if(self.pick1%10 + self.pick2%10 == 0):
                return '장땡'
            else:
                return str(self.pick1%10) + '땡'
        elif(self.pick1 + self.pick2== 31):
            return '38광땡'
        elif(self.pick1 + self.pick2 == 29):
            return '18광땡'
        elif(self.pick1 + self.pick2== 24):
            return '13광땡'
        elif((self.pick1%10 == 1 and self.pick2%10 ==2)or(self.pick1%10 == 2 and self.pick2%10 ==1)):
            return '알리'
        elif((self.pick1%10 == 1 and self.pick2%10 ==4)or(self.pick1%10 == 4 and self.pick2%10 ==1)):
            return '독사'
        elif((self.pick1%10 == 1 and self.pick2%10 ==9)or(self.pick1%10 == 9 and self.pick2%10 ==1)):
            return '구삥'
        elif((self.pick1%10 == 1 and self.pick2%10 ==10)or(self.pick1%10 == 10 and self.pick2%10 ==1)):
            return '장삥'
        elif((self.pick1%10 == 4 and self.pick2%10 ==10)or(self.pick1%10 == 10 and self.pick2%10 ==4)):
            return '장사'
        elif((self.pick1%10 == 4 and self.pick2%10 ==6)or(self.pick1%10 == 6 and self.pick2%10 ==4)):
            return '쎄륙'
        else:
            if((self.pick1 + self.pick2) % 10 == 0):
                return '망통'
            else:
                return str((self.pick1 + self.pick2) % 10) +'끗'

def dual():
    a = deck('a')
    b = deck('b')
    
    a.getPick1()
    b.getPick1()
    bet1 = a.betting()
    bet2 = b.betting()
    if(bet1 != '1' and bet2 != '1'):
        print(a.getResult())
        print(b.getResult())
    
    
dual()
    
```

```python
#3.Implement the remaining simple arithmetic operators (__sub__, __mul__, and __truediv__).
#남은 산술 연산자들을 구현해라. (뺄셈, 곱셋, 나눗셈)
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n
        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        common = gcd(top,bottom)
        self.num = top // common
        self.den = bottom // common
        
    def __str__(self):
        return str(self.num)+"/"+str(self.den)

    def show(self):
        print(self.num,"/",self.den)
        
    def __sub__(self,otherfraction):
        newnum = self.num*otherfraction.den - self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum, newden)
    
    def __mul__(self, otherfraction):
        newnum = self.num*otherfraction.num
        newden = self.den*otherfraction.den
        return Fraction(newnum, newden)
    
    def __truediv__(self, otherfraction):
        newnum = self.num*otherfraction.den
        newden = self.den*otherfraction.num
        return Fraction(newnum, newden)
        
x = Fraction(1,2)
y = Fraction(2,3)
print(x - y)
print(x * y)
print (x / y)
```

```python
#4. Implement the remaining relational operators (__gt__, __ge__, __lt__, __le__, and __ne__)
#남은 관계 연산자들을 구현하라(>, >=, <, <=, !=)
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n
        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        common = gcd(top,bottom)
        self.num = top // common
        self.den = bottom // common
        
    def __str__(self):
        return str(self.num)+"/"+str(self.den)

    def show(self):
        print(self.num,"/",self.den)
        
    def __gt__(self,otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den * otherfraction.num
        return new1 > new2
    
    def __ge__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den * otherfraction.num
        return new1 >= new2
    
    def __lt__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 < new2
    
    def __le__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 <= new2
    
    def __ne__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 != new2
    
x = Fraction(1,2)
y = Fraction(2,3)
print(x > y)
print(x >= y)
print (x < y)
print(x<=y)
print(x!=y)
```

```python
#6. In the definition of fractions we assumed that negative fractions have a negative numerator and a positive denominator. 
#.   Using a negative denominator would cause some of the relational operators to give incorrect results. In general, this is an unnecessary constraint. 
#.   Modify the constructor to allow the user to pass a negative denominator so that all of the operators continue to work properly.
# 분수의 정의에서 음의 분수에는 음수인 분자와 양수인 분모가 있다고 가정했다. 음수인 분모를 사용하게 되면 일부 관계 연산자가 잘못된 결과를 초래한다.
# 일반적으로 이건 불필요한 제약이다. 사용자가 음수인 분모를 전달하더라도 연산자가 제대로 동작을 계속할 수 있도록 생성자를 수정해라!

def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n
        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        if(bottom < 0):
            bottom = bottom * (-1)
            top = top * (-1)
        common = gcd(top,bottom)
        self.num = top // common
        self.den = bottom // common
        
    def __str__(self):
        return str(self.num)+"/"+str(self.den)

    def show(self):
        print(self.num,"/",self.den)
        
    def __gt__(self,otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den * otherfraction.num
        return new1 > new2
    
    def __ge__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den * otherfraction.num
        return new1 >= new2
    
    def __lt__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 < new2
    
    def __le__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 <= new2
    
    def __ne__(self, otherfraction):
        new1 = self.num*otherfraction.den
        new2 = self.den*otherfraction.num
        return new1 != new2

x = Fraction(1,2)
y = Fraction(2,-3)
print(x > y)
print(x >= y)
print (x < y)
print(x<=y)
```

```python
#7. Research the __radd__ method. How does it differ from __add__? When is it used? Implement __radd__.
# __radd_메서드를 연구해라. __add__와 무엇이 다른가? 언제 사용되나? __radd__를 구현해라
# __radd__는 역순 연산자로, a + b 를 b + a 순으로 계산한다.
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n
        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        if(bottom < 0):
            bottom = bottom * (-1)
            top = top * (-1)
        common = gcd(top,bottom)
        self.num = top // common
        self.den = bottom // common
        
    def __str__(self):
        return str(self.num)+"/"+str(self.den)

    def show(self):
        print(self.num,"/",self.den)
        
    def __radd__(self,otherfraction):
        newnum = otherfraction.num*self.den + otherfraction.den * self.num
        newden = otherfraction.den * self.den
        return Fraction(newnum, newden)
x = Fraction(1,2)
y = Fraction(2,3)
print(Fraction.__radd__(x,y))
```

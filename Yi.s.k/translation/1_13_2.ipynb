```python
1.13.2. Inheritance: Logic Gates and Circuits
    상속 : 논리 게이트 및 회로
Our final section will introduce another important aspect of object-oriented programming. 
Inheritance is the ability for one class to be related to another class in much the same way that people can be related to one another.
Children inherit characteristics from their parents. Similarly, Python child classes can inherit characteristic data and behavior from a parent class.
These classes are often referred to as subclasses and superclasses.
마지막 섹션에서는 객체 지향의 중요한 측면을 하나 더 소개시켜주겠다.
상속은 사람이 다른 사람과 연관되는것과 같은 방식으로 한 클래스가 다른 클래스에 연관될 수 있는 기술이다.
자녀는 부모로부터 특성을 상속받는다. 비슷하게, 파이썬 자녀 클래스들은 부모 클래스로부터 특성 데이터나 동작을 상속받을 수 있다.
이 클래스들은 종종 서브클래스와 슈퍼클래스라고 한다.

Figure 8 shows the built-in Python collections and their relationships to one another. We call a relationship structure such as this an inheritance hierarchy. 
For example, the list is a child of the sequential collection. In this case, we call the list the child and the sequence the parent (or subclass list and superclass sequence). 
This is often referred to as an IS-A Relationship (the list IS-A sequential collection). 
This implies that lists inherit important characteristics from sequences, namely the ordering of the underlying data and operations such as concatenation, repetition, and indexing.
그림 8은 내부 파이썬 콜렉션들과 그들의 다른 것과의 관계를 보여준다. 우리는 이와 같은 관계 구조를 상속 계층구조라고 부른다.
예를들어, 목록은 순차 콜렉션의 자녀이다. 이러한 경우에, 우리는 목록을 자녀라고, 순차를 부모라고 부른다.(혹은 서브클래스 목록, 슈퍼클래스 순차)
이건 종종 IS-A관계라고 불린다. (목록은 IS-A 순차 콜렉션이다.)
이는 목록이 순차로부터 중요한 특성, 즉 기본 데이터의 순서 지정과 연결, 반복, 인덱싱과 같은 작업을 상속받음을 의미한다.

Lists, tuples, and strings are all types of sequential collections. They all inherit common data organization and operations. 
However, each of them is distinct based on whether the data is homogeneous and whether the collection is immutable. 
The children all gain from their parents but distinguish themselves by adding additional characteristics.
목록,튜플, 그리고 문자열은 순차 콜렉션의 모든 유형이다. 이들은 모두 공통된 데이터 구성 및 운영을 상속한다.
그러나, 이들 각각은 데이터가 동질적인지 여부와 컬렉션이 불변인지 여부에 따라 구분된다.
아이들은 그들의 부모로부터 모든 것을 받미나 추가적인 특성을 추가함으로서 스스로를 구분시킨다.

By organizing classes in this hierarchical fashion, object-oriented programming languages allow previously written code to be extended to meet the needs of a new situation. 
In addition, by organizing data in this hierarchical manner, we can better understand the relationships that exist. 
We can be more efficient in building our abstract representations.
이런 계층적 방식으로 클래스들을 구성함으로써, 객체 지향 프로그래밍 언어는 새로운 상황의 필요를 충족시키기 위해 이전에 작성된 코드를  확장할 수 있게 한다.
게다가, 데이터를 이러한 계층적 방식으로 구성함으로써, 존재하는 관계들을 더욱 잘 이해할 수 있다.
추상적인 표현을 보다 효율적으로 만들 수 있다.

To explore this idea further, we will construct a simulation, an application to simulate digital circuits. The basic building block for this simulation will be the logic gate. 
These electronic switches represent boolean algebra relationships between their input and their output. In general, gates have a single output line. 
The value of the output is dependent on the values given on the input lines.
이 아이디어를 더 자세히 살펴보기 위해, 디지털 회로를 시뮬레이션하는 시뮬레이션을 구성할 것이다. 이 시뮬레이션의 기본 빌딩 블록은 논리 게이트이다.
이 전자 스위치들은 입출력 간의 불 대수 관계를 표시할 것이다. 일반적으로, 게이트들은 단일 출력 라인을 가지고 있다.
출력 값은 입력 값에 의존한다.

AND gates have two input lines, each of which can be either 0 or 1 (representing False or True, repectively). If both of the input lines have the value 1, the resulting output is 1. 
However, if either or both of the input lines is 0, the result is 0. OR gates also have two input lines and produce a 1 if one or both of the input values is a 1.
In the case where both input lines are 0, the result is 0.
AND 게이트는 0이나 1이 들어올 수 있는 입력 라인이 2개 있다.(각각 거짓과 참을 의미). 만약 입력라인이 둘다 1이면, 출력값도 1이다.
그러나 입력라인 중 하나가, 혹은 입력라인 둘다 0이라면 결과는 0이다. OR 게이트 역시 입력라인이 2개가 있고, 입력라인 중 하나라도 1이 있으면 1을 제공한다.
입력라인이 둘다 0이면, 결과도 0이다.

NOT gates differ from the other two gates in that they only have a single input line. The output value is simply the opposite of the input value.
If 0 appears on the input, 1 is produced on the output. Similarly, 1 produces 0. Figure 9 shows how each of these gates is typically represented. 
Each gate also has a truth table of values showing the input-to-output mapping that is performed by the gate.
NOT 게이트는 단일 입력만 가지고 있다는 점이 다른 두 게이트와 다르다. 출력은 간단히 입력 값의 반대이다.
입력이 0이면, 출력은 1이다. 비슷하게 1이 입력이면 결과는 0이다. 그림 9는 이 게이트들이 어떻게 전형적으로 표시되는지 보여준다.
각 게이트들은 또한 게이트에서 수행되는 입력-to-출력 매핑을 나타내는 진리표도 있다.

By combining these gates in various patterns and then applying a set of input values, we can build circuits that have logical functions. 
Figure 10 shows a circuit consisting of two AND gates, one OR gate, and a single NOT gate. 
The output lines from the two AND gates feed directly into the OR gate, and the resulting output from the OR gate is given to the NOT gate. 
If we apply a set of input values to the four input lines (two for each AND gate), the values are processed and a result appears at the output of the NOT gate. 
Figure 10 also shows an example with values.
이러한 게이트들을 다양한 패턴으로 결합하고 일련의 입력값들을 적용함으로서, 논리적 기능을 가진 회로를 만들 수 있다.
그림 10은 AND게이트 2개, OR 게이트 1개, NOT 게이트 1개로 구성된 회로를 보여준다.
AND 게이트 2개의 출력은 바로 OR 게이트로 공급되며, OR게이트의 결과는 NOT 게이트에 제공된다.
만약 네 개의 입력 라인에 입력 값을 셋팅하면(각 AND게이트에 2개씩), 값이 처리되고 결과가 NOT 게이트의 출력에 나타난다.
그림 10이 값이 있는 예시도 보여주고 있다.

../_images/circuit1.png

In order to implement a circuit, we will first build a representation for logic gates. Logic gates are easily organized into a class inheritance hierarchy as shown in Figure 11. 
At the top of the hierarchy, the LogicGate class represents the most general characteristics of logic gates: namely, a label for the gate and an output line. 
    The next level of subclasses breaks the logic gates into two families, those that have one input line and those that have two. 
    Below that, the specific logic functions of each appear.
회로를 구현하기 위해서는, 먼저 논리 게이트에 대한 표현을 구성해야 한다. 논리 게이트는 그림 11과 같이 간단하게 클래스 상속 계층 구조로 구성된다.
계층 구조의 상단에서, 논리 게이트 클래스는 논리게이트의 가장 일반적인 특성 : 게이트의 라벨과 출력선을 나타낸다.
다음 레벨의 서브클래스들은 논리 게이트들을 두개의 가족으로 나눈다 : 하나는 입력 라인을 한개 가졌고, 다른건 입력 라인을 두개 가졌다.
    그 아래에는 각각에 대한 특정 논리 기능이 나타난다.

../_images/gates.png

We can now start to implement the classes by starting with the most general, LogicGate.
As noted earlier, each gate has a label for identification and a single output line. In addition, we need methods to allow a user of a gate to ask the gate for its label.
이제 가장 일반적ㄷ인 로직게이트부터 클래스를 구현할 수 있다.
먼저 말했듯이, 각 게이트는 식별용 라벨과 단일 출력 게이트가 있다. 추가로, 게이트의 사용자가 라벨에 게이트를 요청하는 것을 허락할 메서드가 필요하다.

The other behavior that every logic gate needs is the ability to know its output value. This will require that the gate perform the appropriate logic based on the current input. 
In order to produce output, the gate needs to know specifically what that logic is. This means calling a method to perform the logic computation. 
The complete class is shown in Listing 8.
모든 로직 게이트가 필요로 하는 동작은 출력 값을 알아내는 능력이다. 이를 위해서는 게이트가 현재 입력을 기반으로 적절한 논리를 수행하여야 한다.
출력을 생성하려면 게이트는 그 로직이 무엇인지 구체적으로 알아야 한다. 이는 논리 연산을 수행하는 메서드를 호출하는것을 의미한다.
전체 클래스는 Listing 8 에서 보여주고 있다.

Listing 8

class LogicGate:

    def __init__(self,n):
        self.label = n
        self.output = None

    def getLabel(self):
        return self.label

    def getOutput(self):
        self.output = self.performGateLogic()
        return self.output
    
At this point, we will not implement the performGateLogic function. The reason for this is that we do not know how each gate will perform its own logic operation. 
Those details will be included by each individual gate that is added to the hierarchy. This is a very powerful idea in object-oriented programming. 
We are writing a method that will use code that does not exist yet. The parameter self is a reference to the actual gate object invoking the method. 
Any new logic gate that gets added to the hierarchy will simply need to implement the performGateLogic function and it will be used at the appropriate time. 
Once done, the gate can provide its output value. This ability to extend a hierarchy that currently exists and provide the specific functions that the hierarchy needs to use the new class is extremely important for reusing existing code.
여기서, 우리는 performGateLogic 함수를 구현하지 않을 것이다. 각 게이트가 스스로의 로직 연산을 수행할지 말지를 알 수 없기 때문이다.
저 세부정보들은 계층에 추가된 각 개별 게이트에 포함된다. 이건 객체 지향 프로그래밍에서 매우 강력한 생각이다.
우리는 아직 존재하지 않는 코드를 사용할 메서드를 작성하고 있다. 파라미터 self는 메서드를 호출하는 실제 게이트 객체에 대한 참조이다.
계층에 추가된 모든 새로운 논리 게이트는 performGateLogic 함수를 구현해서 적절한 시간에 사용되는 것이 필요하다.
한번 완료되면, 게이트는 출력 값을 제공할 수 있다. 계층이 새로운 클래스를 사용하기 위해 필요한 현재 존재하는 계층을 확장하고 특정 함수를 제공하는 이 기능은 존재하는 코드를 재사용 하는데에 매우 중요하다.

We categorized the logic gates based on the number of input lines. The AND gate has two input lines. The OR gate also has two input lines. 
NOT gates have one input line. The BinaryGate class will be a subclass of LogicGate and will add two input lines. 
The UnaryGate class will also subclass LogicGate but will have only a single input line. In computer circuit design, these lines are sometimes called “pins” so we will use that terminology in our implementation.
우리는 입력 라인의 수에 따라 로직 게이트를 분류했다. AND 게이트는 입력 라인 두개를 가졌다. OR 게이트 역시 입력라인을 두개 가졌다.
NOT 게이트는 입력 라인을 하나 가지고 있다. BinaryGate 클래스는 LogicGate의 서브클래스가 될 것이고 입력 라인 두개를 추가할 것이다.
UnaryGate 클래스 또한 LogicGate의 서브클래스 이지만 단일 입력 라인을 가질 것이다. 컴퓨터 회로 디자인에서, 이런 라인은 떄떄로 핀이라고 불리고 우리는 이 전문 용어를 사용할 것이다.

Listing 9

class BinaryGate(LogicGate):

    def __init__(self,n):
        LogicGate.__init__(self,n)

        self.pinA = None
        self.pinB = None

    def getPinA(self):
        return int(input("Enter Pin A input for gate "+ self.getLabel()+"-->"))

    def getPinB(self):
        return int(input("Enter Pin B input for gate "+ self.getLabel()+"-->"))
Listing 10

class UnaryGate(LogicGate):

    def __init__(self,n):
        LogicGate.__init__(self,n)

        self.pin = None

    def getPin(self):
        return int(input("Enter Pin input for gate "+ self.getLabel()+"-->"))
    
Listing 9 and Listing 10 implement these two classes. The constructors in both of these classes start with an explicit call to the constructor of the parent class using the parent’s __init__ method.
When creating an instance of the BinaryGate class, we first want to initialize any data items that are inherited from LogicGate.
In this case, that means the label for the gate. The constructor then goes on to add the two input lines (pinA and pinB). 
This is a very common pattern that you should always use when building class hierarchies. 
Child class constructors need to call parent class constructors and then move on to their own distinguishing data.
Listing 9와 Listing 10에선 이 두 클래스를 구현한다. 이 두 클래스의 생성자들은 부모의 _init_ 메서드를 사용하여 부모 클래스의 생성자에 대한 명시적 호출로 시작한다.
BinaryGate 클래스의 인스턴스를 만들 때, 먼저 LogicGate에서 상속된 데이터 항목을 초기화 하길 원한다.
이 경우에, 게이트의 라벨을 의미한다. 그리고 생성자는 두 입력 라인을 추가한다.
이는 클래스 계층 구조르 ㄹ 만들 때에 항상 사용해야 하는 매우 흔한 패턴이다.
자식 클래스 생성자는 부모 클래스 생성자를 호출해서 고유한 구분 데이터로 이동해야 한다.

Python also has a function called super which can be used in place of explicitly naming the parent class. 
This is a more general mechanism, and is widely used, especially when a class has more than one parent. 
But, this is not something we are going to discuss in this introduction.
For example in our example above LogicGate.__init__(self,n) could be replaced with super(UnaryGate,self).__init__(n).
파이썬은 부모 클래스를 명시적으로 이름짓는 대신 사용할 수 있는 super라는 함수를 가지고 있다.
이건 좀더 일반적이고 널리 사용되는 매커니즘인데, 특히 클래스의 부모가 하나 초과일때 그렇다.
하지만, 이건 이번 소개에서 논의할 내용이 아니다.
예를들어 위의 예시에서 LogicGate._init_(self , n)은 super(UnartGate,self)._init_(n)으로 대체될 수 있다.

The only behavior that the BinaryGate class adds is the ability to get the values from the two input lines.
Since these values come from some external place, we will simply ask the user via an input statement to provide them. 
The same implementation occurs for the UnaryGate class except that there is only one input line.
BinaryGate 클래스가 추가하는 유일한 동작은 두 입력 라인으로부터 값을 가져오는 기능이다.
이 값들은 외부에서 왔으므로, 사용자에게 입력 문을 통해 제공하도록 요청할 것이다.
단일 입력 라인이 있다는 것을 제외하고 같은 실행이 UnaryGate 클래스에서 일어난다.

Now that we have a general class for gates depending on the number of input lines, we can build specific gates that have unique behavior.
For example, the AndGate class will be a subclass of BinaryGate since AND gates have two input lines.
As before, the first line of the constructor calls upon the parent class constructor (BinaryGate), which in turn calls its parent class constructor (LogicGate). 
Note that the AndGate class does not provide any new data since it inherits two input lines, one output line, and a label.
입력 라인의 수에 의존하는 게이트에 대한 일반 클래스가 있으니, 고유한 동작을 하는 특정 게이트를 만들 수 있다.
예를들어, AndGate 클래스는 AND 게이트가 입력 라인 두개를 가지고 있기 떄문에 BinaryGate의 서브클래스가 된다.
이전처럼, 생성자의 첫번째 라인은 부모 클래스 생성자(LogicGate)를 호출하는 부모 클래스 생성자(BinaryGate)를 호출한다.
AndGate 클래스는 입력 라인 두개, 출력 라인 한개, 라벨을 상속하므로 새로운 데이터를 제공하지 않는다.

Listing 11

class AndGate(BinaryGate):

    def __init__(self,n):
        BinaryGate.__init__(self,n)

    def performGateLogic(self):

        a = self.getPinA()
        b = self.getPinB()
        if a==1 and b==1:
            return 1
        else:
            return 0
The only thing AndGate needs to add is the specific behavior that performs the boolean operation that was described earlier. 
This is the place where we can provide the performGateLogic method. 
For an AND gate, this method first must get the two input values and then only return 1 if both input values are 1. 
The complete class is shown in Listing 11.
AndGate가 추가해야하는 유일한 것은 일찍이 설명한 부울 연산을 수행하는 특정 동작이다.
여긴 performGateLogic 메서드를 제공할 수 있느 ㄴ곳이다.
AND게이트의 경우, 이 메서드는 먼저 입력 값 두개를 가져와야 하며 입력 값 두개가 1일때에만 1을 반환한다.
전체 클래스는 Listing 11에서 보여주고 있다.

We can show the AndGate class in action by creating an instance and asking it to compute its output. 
The following session shows an AndGate object, g1, that has an internal label "G1". 
When we invoke the getOutput method, the object must first call its performGateLogic method which in turn queries the two input lines. 
Once the values are provided, the correct output is shown.
인스턴스를 생섣ㅇ하고, 출력을 계산하도록 요구함으로서 동작하는 AndGate 클래스를 보여줄 수 있다.
다음 세션에서는 내부 라벨 G1을 갖고 있는 AndGate 객체 g1을 보여주고 있다.
getOutput 메서드를 호출할 때, 객체는 반드시 먼저 두 입력 라인을 차례로 조회하는 performGateLogic 메서드를 호출해야 한다.
한번 값이 제공되면, 정확한 출력이 보여진다.

>>> g1 = AndGate("G1")
>>> g1.getOutput()
Enter Pin A input for gate G1-->1
Enter Pin B input for gate G1-->0
0

The same development can be done for OR gates and NOT gates. 
The OrGate class will also be a subclass of BinaryGate and the NotGate class will extend the UnaryGate class. 
Both of these classes will need to provide their own performGateLogic functions, as this is their specific behavior.
OR 게이트와 NOT 게이트에서도 동일한 개발이 가능하다.
OrGate 클래스 역시 BinaryGate의 서브클래스가 될 것이고, NotGate는 UnaryGate 클래스를 확장할 것이다.
이 클래스들 둘 다 고유한 performGateLogic 함수를 제공해야 한다. 이것이 그들의 구체적인 행동이기 때문이다.

We can use a single gate by first constructing an instance of one of the gate classes and then asking the gate for its output (which will in turn need inputs to be provided). For example:
게이트 클래스 중 하나의 인스턴스를 첫 생성하고 게이트에 출력을 요구함으로서 단일 게이트를 사용할 수 있다. 예시 : 
    
>>> g2 = OrGate("G2")
>>> g2.getOutput()
Enter Pin A input for gate G2-->1
Enter Pin B input for gate G2-->1
1
>>> g2.getOutput()
Enter Pin A input for gate G2-->0
Enter Pin B input for gate G2-->0
0
>>> g3 = NotGate("G3")
>>> g3.getOutput()
Enter Pin input for gate G3-->0
1
Now that we have the basic gates working, we can turn our attention to building circuits. 
In order to create a circuit, we need to connect gates together, the output of one flowing into the input of another. 
To do this, we will implement a new class called Connector.
이제 기본 게이트가 작동했으니, 회로 구성에 집중할 수 있다.
회로를 만드려면, 게이트를 서로 연결해야 하는데, 하나의 출력이 다른 하나의 입력으로 흘러들어간다.
이렇게 하려면 Connector라고 불리는 새로운 클래스를 구현해야 한다.

The Connector class will not reside in the gate hierarchy. It will, however, use the gate hierarchy in that each connector will have two gates, one on either end (see Figure 12). 
This relationship is very important in object-oriented programming. It is called the HAS-A Relationship. 
Recall earlier that we used the phrase “IS-A Relationship” to say that a child class is related to a parent class, for example UnaryGate IS-A LogicGate.
Connector 클래스는 게이트 계층에 상주하지 않을 것이다. 그러나, 각 connector는 두개씩 가지고 있는 게이트의 각 끝에 있는 게이트 계층을 사용할 것이다.( 그림 12)
이 관계는 객체 지향 프로그래밍에서 매우 중요하다. HAS-A 관계라고 불린다.
이전에 자녀 클래스는 부모 클래스와 관련이 있다고 말하기 위해 사용햇던 IS-A 관계를 상기해 봐라, 예를들어 UnartGate IS-A LogicGate.

../_images/connector.png

Now, with the Connector class, we say that a Connector HAS-A LogicGate meaning that connectors will have instances of the LogicGate class within them but are not part of the hierarchy. 
When designing classes, it is very important to distinguish between those that have the IS-A relationship (which requires inheritance) and those that have HAS-A relationships (with no inheritance).
이제, Connector는 connector가 LogicGate 클래스의 인스턴스를 갖고 있지만 계층의 일부는 아니다는 의미인 Connector HAS-A Logic Gate라고 말할 것이다.
클래스를 디자인할때, 상속이 필요한 IS-A 관계를 가진 것과 그렇지 않은 HAS-A 관계를 가진 것을 구분하는 것이 매우 중요하다.

Listing 12 shows the Connector class. The two gate instances within each connector object will be referred to as the fromgate and the togate, recognizing that data values will “flow” from the output of one gate into an input line of the next. 
The call to setNextPin is very important for making connections (see Listing 13). We need to add this method to our gate classes so that each togate can choose the proper input line for the connection.
Listing 12는 Connector class를 보여준다. 각 connector 객체 안의 두 게이트 인스턴스들은 데이터 값이 한 게이트의 입력 라인에서 다음 것의 입력 라인으로 흐름을 인지하여 fromgate와 togate로 참조된다.
setNextPin에 대한 호출은 연결을 만드는 데에 매우 중요하다. 이 메소드를 게이트 클래스에 추가하여 각 togate가 연결에 적절한 입력 라인을 선택할 수 있도록 해줄 필요가 있다.

Listing 12

class Connector:

    def __init__(self, fgate, tgate):
        self.fromgate = fgate
        self.togate = tgate

        tgate.setNextPin(self)

    def getFrom(self):
        return self.fromgate

    def getTo(self):
        return self.togate
In the BinaryGate class, for gates with two possible input lines, the connector must be connected to only one line.
If both of them are available, we will choose pinA by default. If pinA is already connected, then we will choose pinB.
It is not possible to connect to a gate with no available input lines.
BinaryGate 클래스에서, 사용 가능한 두 입력 라인이 있는 게이트에는 connector가 하나의 라인에만 연결되어야 한다.
둘다 사용할 수 있다면 기본으로 pinA를 선택한다. 만약 pinA가 이미 연결되어 있다면, pinB를 고른다.
사용 가능한 입력 라인이 없는 게이트에는 연결할 수 없다.

Listing 13

def setNextPin(self,source):
    if self.pinA == None:
        self.pinA = source
    else:
        if self.pinB == None:
            self.pinB = source
        else:
           raise RuntimeError("Error: NO EMPTY PINS")
Now it is possible to get input from two places: externally, as before, and from the output of a gate that is connected to that input line. 
    This requires a change to the getPinA and getPinB methods (see Listing 14). If the input line is not connected to anything (None), then ask the user externally as before. 
    However, if there is a connection, the connection is accessed and fromgate’s output value is retrieved. This in turn causes that gate to process its logic. 
    This continues until all input is available and the final output value becomes the required input for the gate in question. 
    In a sense, the circuit works backwards to find the input necessary to finally produce output.
이제 두 장소로부터 입력을 받을 수 있다 : 외부, 입력 라인에 연결된 게이트의 출력.
이렇게 하려면 getPinA와 getPinB 메서드를 바꿔야 한다. 만약 입력 라인이 아무곳에도 연결되어 있지 않다면 이전처럼 사용자에게 외부적ㄷ으로 요구해라.
그러나 만약 연결이 있으면, 연결은 접속되고 fromgate의 출력 값도 검색된다. 이것은 게이트가 차례로 로직을 처리하도록 만든다.
모든 입력을 사용할 수 있고 마지막 출력 값이 요구된 게이트에 필요한 입력이 될 때까지 계속된다.
어떤 면에선, 회로가 마지막으로 출력을 제공하는데에 필요한 입력을 찾기 위해 거꾸로 동작한다.

Listing 14

def getPinA(self):
    if self.pinA == None:
        return input("Enter Pin A input for gate " + self.getName()+"-->")
    else:
        return self.pinA.getFrom().getOutput()
The following fragment constructs the circuit shown earlier in the section:
다음 단락에서는 전에 보여준 회로를 구성한다.

>>> g1 = AndGate("G1")
>>> g2 = AndGate("G2")
>>> g3 = OrGate("G3")
>>> g4 = NotGate("G4")
>>> c1 = Connector(g1,g3)
>>> c2 = Connector(g2,g3)
>>> c3 = Connector(g3,g4)

The outputs from the two AND gates (g1 and g2) are connected to the OR gate (g3) and that output is connected to the NOT gate (g4). 
The output from the NOT gate is the output of the entire circuit. For example:
AND 게이트 두 개의 출력은(g1과 g2) OR게이트에 연결되고(g3) 그 출력은 NOT 게이트에 연결된다.(g4)
NOT 게이트의 출력이 전체 회로의 출력이다. 예를 들어 : 

>>> g4.getOutput()
Pin A input for gate G1-->0
Pin B input for gate G1-->1
Pin A input for gate G2-->1
Pin B input for gate G2-->1
0
Try it yourself using ActiveCode 4.

RunLoad HistoryShow CodeLens

1
class LogicGate:
2
​
3
    def __init__(self,n):
4
        self.name = n
5
        self.output = None
6
​
7
    def getName(self):
8
        return self.name
9
​
10
    def getOutput(self):
11
        self.output = self.performGateLogic()
12
        return self.output
13
​
14
​
15
class BinaryGate(LogicGate):
16
​
17
    def __init__(self,n):
18
        LogicGate.__init__(self,n)
19
​
20
        self.pinA = None
21
        self.pinB = None
22
​
23
    def getPinA(self):
24
        if self.pinA == None:
25
            return int(input("Enter Pin A input for gate "+self.getName()+"-->"))
26
        else:
27
            return self.pinA.getFrom().getOutput()
28
​
29
    def getPinB(self):
30
        if self.pinB == None:
31
            return int(input("Enter Pin B input for gate "+self.getName()+"-->"))
32
        else:
33
            return self.pinB.getFrom().getOutput()
34
​
The Complete Circuit Program. (complete_cuircuit)
Self Check
자가 점검
Create a two new gate classes, one called NorGate the other called NandGate. NandGates work like AndGates that have a Not attached to the output. 
NorGates work lake OrGates that have a Not attached to the output.
게이트 두개를 새로 만들자. 하나는 NorGate, 다른 하나는 NandGate. NandGate는 출력에 Not이 붙은 AndGate처럼 동작한다.
NorGate는 출력에 Not이 붙은 OrGate처럼 동작한다.

class NandGate(BinaryGate):

    def __init__(self,n):
        BinaryGate.__init__(self,n)

    def performGateLogic(self):

        a = self.getPinA()
        b = self.getPinB()
        if a==1 and b==1:
            return 0
        else:
            return 1

class NorGate(BinaryGate):

    def __init__(self,n):
        BinaryGate.__init__(self,n)

    def performGateLogic(self):

        a = self.getPinA()
        b = self.getPinB()
        if a ==1 or b==1:
            return 0
        else:
            return 1

Create a series of gates that prove the following equality NOT (( A and B) or (C and D)) is that same as NOT( A and B ) and NOT (C and D). 
Make sure to use some of your new gates in the simulation.
다음 두 개의 등식ㄷ이 동일하다는 것을 증명하는 일련의 게이트를 만들어라.
시뮬레이션의 새 게이트 중 일부를 꼭 사용해야 한다.
```
